```python
import numpy.random as rnd
import numpy as np
from matplotlib import pyplot as plt
# Your solution 
N_total = [10**i for i in range(1, 6)]
N_green = 0
repeat = 100
def estimate_pi(n):
    x = rnd.rand(n)
    y = rnd.rand(n)
    N_green = np.sum((x**2 + y**2) <= 1)
    return (4*N_green/n)
mean =[]
standard_deviation = []

for i in N_total:
    temp = [estimate_pi(i) for j in range(repeat)]
    mean.append(np.mean(temp))
    standard_deviation.append(np.std(temp))
    
x_values = N_total
plt.errorbar(x_values, mean, yerr = standard_deviation, fmt = 'o', ecolor = 'b', label = 'Std. Dev. of $\pi$ values')
plt.xscale('log')
plt.axhline(y= np.pi, label = 'Real $\pi$', ls = 'dashed')
plt.title('Value of $\pi$ vs. No. of Points Used')
plt.ylabel('Value of $\pi$')
plt.xlabel('No. of points used to generate value')
plt.xticks(x_values)
plt.legend()
plt.show()
```
